We want to calculate the angular motion of the following velocity field:

\begin{equation}
\mathbf{v} = (x^2 - y^2)\mathbf{i} - 2xy \mathbf{j}
\end{equation}

Then, we will decide if the flow is rotational or not.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate Grid
x = np.linspace(-2,2,50)
y = np.linspace(-2,2,50)

xx, yy = np.meshgrid(x, y, indexing='xy')

# Calculate velocity at the grid points
vx = (xx**2 - yy**2)
vy = -2*xx*yy

# Plot the streamlines
fig, axs = plt.subplots(1, 2,figsize=(12,7))
axs = axs.flat

#  Varying density along a streamline
axs[0].streamplot(xx, yy, vx, vy, density=[0.5, 1])
axs[0].set_title('VELOCITY')
```

Let's calculate the angular motion in each direction. The velocity components are:
\begin{equation} 
u = x^2 - y^2, \,\, v = -2xy, \,\, w = 0
\end{equation}

Then the angular rotation in each direction is
\begin{equation}
\omega_x = \frac{1}{2}\left(\frac{\partial w}{\partial y} - \frac{\partial v}{\partial z}\right) = 0
\end{equation}

\begin{equation}
\omega_y = \frac{1}{2}\left(\frac{\partial u}{\partial z} - \frac{\partial w}{\partial x}\right) = 0
\end{equation}

\begin{equation}
\omega_z = \frac{1}{2}\left(\frac{\partial v}{\partial x} - \frac{\partial u}{\partial y}\right) = \frac{1}{2}(-2y - (-2y)) = 0
\end{equation}
